# Exploratory Analysis

## Revenue and Order Size

In Figure 1, we observed that the 'email' sales group generated the highest revenue, followed by 'email + call.' This aligns with Figure 2 showing that this group had the most customers.

Figure 9 shoes that for the 'email' group, as well as in general, orders of 9, 10, and 11 brought in the highest revenue. However, for orders of 12 items or more, the 'email' group's revenue decreased, and it was replaced by the 'email + call' group.

Figure 10 shows an overall upward trend, indicating that orders with a greater number of products generated higher revenue. On average, customers in the 'email+call' group spent more per order than the other groups. It is worth noting that the 'call' group had no orders exceeding 14 items.

## Revenue Over Time

Customers who were in the 'email' group were emailed when the product line was launched, followed up by email three weeks later. 

Figure 11 shows that the week with the highest revenue is week 1, and the method that generated the most revenue that week was 'email', which correlates with the email sent at launch. There is then a decrease in revenue for this method until week 4, when it shows another minor spike, which is consistent with the follow-up email sent at that time.

In contrast, revenue from customers in the 'call' group remained consistently low, without much of a discernible pattern. Although it is known that these calls took approximately 30 minutes per customer, it is unclear whether they were all completed in one week or spaced out.

Customers in the 'email and call' group were sent an email at launch, followed by a phone call a week later. It is logical that revenue was initially low in the first week because there was insufficient time for the follow-up email. This method's revenue then steadily increased over the first five weeks as customers received their email, had time to think, and then received a phone call. In weeks 5 and 6, more sales were generated from this group than any other.

# Hybrid Approach

After completing my analysis I would my recommend  adopting a hybrid approach to marketing. 
Based on my analysis, the 'email' group generated the highest overall spending. However, the 'email + call' group spends the most on average and continues to spend the most in later weeks. 
My recommendation is to send out an email during the first week of the launch and then follow up with a call around week 4 for customers who have not yet made a purchase. This will allow us to capitalise on the strengths of both groups and improve our overall revenue.

# Business Metric to Monitor
